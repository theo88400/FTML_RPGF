## Question 2
Proof that $A^TA$ for square matrices is a double sum of the squared elements of $A$.
Let $(A,B)\in\mathbb{R}^{n\times n}\times\mathbb{R}^{n\times n}$, we have for a product :
$
\begin{equation}
\begin{split}
(AB)_{ij}=&\sum_{k=1}^nA_{ik}B_{kj}\\
\end{split}
\end{equation}
$
Therefore for $A^TA$ :
$
\begin{equation}
\begin{split}
(A^TA)_{ij}=&\sum_{k=1}^nA^T_{ik}A_{kj}\\
=&\sum_{k=1}^nA_{ki}A_{kj}\\
=&\sum_{k=1}^nA_{ik}A_{jk}\\
\end{split}
\end{equation}
$
For the diagonal elements, we have :
$
\begin{equation}
\begin{split}
tr(A^TA)=&\sum_{i=1}^n\sum_{k=1}^nA_{ik}A_{ki}\\
=&\sum_{i=1}^nA_{ii}^2\\
\end{split}
\end{equation}
$

## Question 3
Proof that $E_{\epsilon}\left[\frac{1}{n}\|A\epsilon\|_{2}^2\right]=\frac{\sigma^2}{n}\text{tr}(A^TA)$.

1. Starting with the left-hand side:

$E_{\epsilon}\left[\frac{1}{n}\|A\epsilon\|_{2}^2\right]=E_{\epsilon}\left[\frac{1}{n}\sum_{i=1}^n(A\epsilon)_i^2\right].$

2. Expanding $(A\epsilon)_i$, which is the i-th row of matrix $A$ times vector $\epsilon$:

$=E_{\epsilon}\left[\frac{1}{n}\sum_{i=1}^n\left(\sum_{k=1}^nA_{ik}\epsilon_k\right)^2\right].$

3. Expanding the square inside the sum, we have

$E_{\epsilon}\left[\frac{1}{n}\sum_{i=1}^n\left(\sum_{k=1}^n A_{ik}\epsilon_k\right)\left(\sum_{j=1}^n A_{ij}\epsilon_j\right)\right]$

4. Using the linearity of expectation and the distributive property of summation, we can interchange the expectation and summation:

$\frac{1}{n}\sum_{i=1}^n\sum_{k=1}^n\sum_{j=1}^n A_{ik}A_{ij}E_{\epsilon}[\epsilon_k\epsilon_j]$

Now, notice that the term inside the expectation, $E_{\epsilon}[\epsilon_k\epsilon_j]$, is non-zero only when $k = j$ (because the $\epsilon_k$ terms are independent and have mean $0$), and in this case, it is equal to the variance of $\epsilon_k$, which is $\sigma^2$. Since these variables are independant, $E[AB]=E[A]E[B]$.
 Therefore, we can rewrite the above expression as:

$=\frac{1}{n}\sum_{i=1}^n\sum_{k=1}^nA_{ik}^2E_{\epsilon}\left[\epsilon_k^2\right].$

5. We know that the variance of $\epsilon_k$ is $\sigma^2$, and since $\epsilon_k$ is centered and because $Var(\epsilon_k)=E[\epsilon_k^2]-E[\epsilon_k]^2$, we get that $E[\epsilon_k^2] = \sigma^2$. Using this, we get:

$=\frac{1}{n}\sum_{i=1}^n\sum_{k=1}^nA_{ik}^2\sigma^2$

6. Taking $\sigma^2$ outside the sum:

$=\frac{\sigma^2}{n}\sum_{i=1}^n\sum_{k=1}^nA_{ik}^2.$

7. The double summation of $A_{ik}^2$ is actually the trace of $A^TA$, so:

$=\frac{\sigma^2}{n}\text{tr}(A^TA).$

## Question 4

Show that $A^TA=A$
We know that :
$
\begin{equation}
\begin{split}
A=&I_n - X(X^TX)^{-1}X^T\\
\end{split}
\end{equation}
$
Let $M=X(X^TX)^{-1}X^T$, we have :
$
\begin{equation}
\begin{split}
A^TA=&(I_n - M)^T(I_n - M)\\
\end{split}
\end{equation}
$
Since $M$ is symmetric, because
$
\begin{equation}
\begin{split}
M^T=(X(X^TX)^{-1}X^T)^T=(X^T)^T((X^TX)^{-1})^TX^T=X(X^TX)^{-1}X^T=M\\
\end{split}
\end{equation}
$
We have :
$
\begin{equation}
\begin{split}
=&(I_n - M)(I_n - M)\\
=&I_n - M - M + M^2\\
\end{split}
\end{equation}
$
In fact, $M^2=M$ because
$
\begin{equation}
\begin{split}
M^2=&(X(X^TX)^{-1}X^T)(X(X^TX)^{-1}X^T)\\
=&X(X^TX)^{-1}(X^TX)(X^TX)^{-1}X^T\\
=&X(X^TX)^{-1}X^T\\
=&M\\
\end{split}
\end{equation}
$
Therefore :
$
\begin{equation}
\begin{split}
=&I_n - M - M + M\\
=&I_n - M\\
=&A\\
\end{split}
\end{equation}
$

## Question 5
Conclude

We have :
$
\begin{equation}
\begin{split}
E_\epsilon[\frac{1}{n}||A\epsilon||_2^2]=&\frac{\sigma^2}{n}tr(A^TA)\\
=&\frac{\sigma^2}{n}tr(A)\\
=&\frac{\sigma^2}{n}tr(I_n - X(X^TX)^{-1}X^T)\\
=&\frac{\sigma^2}{n}(tr(I_n) - tr(X(X^TX)^{-1}X^T))\\
=&\frac{\sigma^2}{n}(n - tr(X(X^TX)^{-1}X^T))\\
\end{split}
\end{equation}
$
However, trace is invariant by cyclic permutation, which means $tr(ABC)=tr(BCA)=tr(CAB)$. Therefore :
$
\begin{equation}
\begin{split}
=&\frac{\sigma^2}{n}(n - tr(X^TX(X^TX)^{-1}))\\
=&\frac{\sigma^2}{n}(n - tr(I_d))\\
=&\frac{\sigma^2}{n}(n - d)\\
\end{split}
\end{equation}
$

## Question 6
Still in the same setting, what is the expected value of $\frac{||y-X\hat\theta||_2^2}{n-d}$?

This equation is the mean square error of the estimator $\hat\theta$, normalized by the degrees of freedom $n-d$.

We know that the expected value of the MSE if $nR_X(\hat\theta)$ because it's $n$ times the number of observations :

And since
$
\begin{equation}
\begin{split}
E(R_X(\hat\theta))=&\frac{\sigma^2}{n}(n - d)\\
\end{split}
\end{equation}
$

We replace it in the value $\frac{||y-X\hat\theta||_2^2}{n-d}$ and we have :
$
\begin{equation}
\begin{split}
E(\frac{||y-X\hat\theta||_2^2}{n-d})=&\frac{E(||y-X\hat\theta||_2^2)}{n-d}\\
=&\frac{E(nR_X(\hat\theta))}{n-d}\\
=&\frac{n}{n-d}E(R_X(\hat\theta))\\
\end{split}
\end{equation}
$
We replace the value of $E(R_X(\hat\theta))$ and we have :
$
\begin{equation}
\begin{split}
=&\frac{n}{n-d}\frac{\sigma^2}{n}(n - d)\\
=&\sigma^2\\
\end{split}
\end{equation}
$

## Question 7

```python
# Produce a numerical simulation that estimates $\sigma^2$
# thanks to the result of question 6. Check that the result is consistent with the theoretical value you have chosen.

from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
n = 10000
d = 2
for sigma in [2, 4, 8, 16, 32]:
    X = np.random.normal(0, 1, (n, d))
    theta = np.arange(1, d + 1)
    Y = X @ theta + np.random.normal(0, sigma, n)

    # Compute the OLS estimator
    reg = linear_model.LinearRegression()
    reg.fit(X, Y)
    theta_hat = reg.coef_
    sigma_hat_squared = np.sum((Y - X @ theta_hat)**2) / (n - d)

    # Print the results
    print("Theoretical value of sigma²: ", sigma**2)
    print("Estimated value of sigma²: ", sigma_hat_squared)

    # Plot the results
    plt.scatter(X[:, 0], Y, color='black')
    plt.scatter(X[:, 0], X @ theta_hat, color='blue')
    plt.show()
```
